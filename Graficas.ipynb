#Punto 1


```python

import matplotlib.pyplot as plt

# Funcion que genera Fibonacci
def generar_fibonacci(n):
    fib = []
    a = 0
    b = 1
    for num in range(n):
        fib.append(a)
        valores = [a, b]
        a = valores[1]
        b = valores[0] + valores[1]
    return fib

# Entrada del usuario
n = int(input("Ingrese el número de términos de Fibonacci: "))


x=range(n)
y= generar_fibonacci(n)

# Configuracion del grafico
plt.figure(figsize=(10, 6))
plt.grid(color='k')
plt.plot(x, y, '*--k')
plt.title(f'Primeros {n} términos de Fibonacci')
plt.xlabel('Posición (n)')
plt.ylabel('Valor')
plt.xticks(x)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


#Datos
temperaturas = [25.6, 26.1, 24.8, 27.3, 28.4, 26.7, 25.9, 29.1, 30.2, 27.8, 24.3, 23.9,
                26.5, 28.0, 27.4, 29.3, 30.5, 26.8, 25.7, 24.6]

#Calculo de media, mediana y desviacion
media = np.mean(temperaturas)
mediana = np.median(temperaturas)
desviacion_estandar = np.std(temperaturas)

print("Resultados: ")
print(f"Media: {media:.2f}")
print(f"Mediana: {mediana:.2f}")
print(f"Desviación estándar: {desviacion_estandar:.2f}")


x = list(range(1, len(temperaturas) + 1))
y = temperaturas

#Grafico
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'o-b')
plt.xlabel('Meses')
plt.ylabel('Temperatura (°C)')
plt.title('Temperatura media mensual')



plt.show()
```

```python
import random
import matplotlib.pyplot as plt

#Simulacion de 1000 lanzamientos
lanzamientos = [random.randint(1, 6) for j in range(1000)]

#Contar la frecuencia de cada numero
frecuencia = []
for i in range(1, 7):
    frecuencia.append(lanzamientos.count(i))


x = range(1, 7)
y = frecuencia
#Creacion del grafico
plt.bar(x, y, color='y', edgecolor='k')
plt.xlabel('Número del dado')
plt.ylabel('Frecuencia')
plt.title('Simulación de 1000 lanzamientos de un dado')
plt.xticks(range(1, 7))

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar valores de x de 0 a 2π
x = np.linspace(0, 2 * np.pi, 1000)

# Calcular las funciones trigonométricas
y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(x)

#Crear figura y ejes
plt.figure(figsize=(10, 6))


# Graficar seno, coseno y tangente
plt.plot(x, y_sin, label="Seno", color="b")
plt.plot(x, y_cos, label="Coseno", color="r")
plt.plot(x, y_tan, label="Tangente", color="g", linestyle = "--")
plt.ylim(-5, 5)  # Limitar el eje Y

# Configuracion del grafico
plt.axhline(0, color='k', linewidth=1)  # Eje X
plt.axvline(0, color='k', linewidth=1)  # Eje Y
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi], ['0', 'π/2', 'π', '3π/2', '2π'])
plt.legend()
plt.title("Funciones Trigonométricas: Seno, Coseno y Tangente")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def graficar_raices():
    x = np.linspace(0, 10, 100)
    y_sqrt = np.sqrt(x)
    y_line = x

    plt.plot(x, y_sqrt, label='y = √x', color='blue')
    plt.plot(x, y_line, label='y = x', color='red', linestyle='dashed')

    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Gráfica de y = √x y y = x')
    plt.legend()
    plt.grid()
    plt.show()

graficar_raices()
```

```python
import numpy as np
import matplotlib.pyplot as plt
def graficar_pendiente(x1, y1, x2, y2):
    m = (y2 - y1) / (x2 - x1)
    x = np.array([x1, x2])
    y = np.array([y1, y2])

    plt.plot(x, y, marker='o', linestyle='-', color='green', label=f'Recta (m={m})')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Gráfica de la recta')
    plt.legend()
    plt.grid()
    plt.show()
graficar_pendiente(1,2,4,8)
```

#Punto 7

```python
import numpy as np
import matplotlib.pyplot as plt

def trayectoria(angulo, VelocidadInicial):
    g = 10
    angulo = np.radians(angulo)
    t_vuelo = 2 * VelocidadInicial * np.sin(angulo) / g
    t = np.linspace(0, t_vuelo, num=100)
    x = VelocidadInicial * np.cos(angulo) * t
    y = VelocidadInicial * np.sin(angulo) * t - 0.5 * g * t**2
    return x, y

angulos = [30, 45, 60]
for angulo in angulos:
    x, y = trayectoria(angulo, 20)
    plt.plot(x, y, label=f'{angulo}°')

plt.xlabel("Distancia")
plt.ylabel("Altura")
plt.title("Movimiento Parabolico")
```

```python
# @title Control deslizante para funciones potencia { run: "auto" }
import numpy as np
import matplotlib.pyplot as plt

n = 7 # @param {"type":"slider","min":1,"max":10,"step":1}

# Configuración inicial
plt.figure(figsize=(10, 6))
x = np.linspace(-2, 2, 400)
plt.clf()  # Limpiar gráfico anterior

# Generar todas las funciones hasta el valor de n
for i in range(1, n + 1):
    y = x**i
    plt.plot(x, y, label=f'$x^{i}$', alpha=0.7, linewidth=2)

#Grafico
plt.title(f'Funciones: $f(x) = x^n$ (n = 1 a {n})')
plt.xlabel('x', fontsize=12)
plt.ylabel('$f(x)$', fontsize=12)
plt.ylim(-5, 5)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(loc='upper left')
plt.show()
```

#Punto 9


```python
import numpy as np
import matplotlib.pyplot as plt
import random

v1 = np.array([0, 0])
v2 = np.array([1, 0])
v3 = np.array([0.5, np.sqrt(3)/2])

puntos_x = []
puntos_y = []

punto = np.array([random.uniform(0, 1), random.uniform(0, 1)])


iteraciones = 5000
contador = 0

while contador < iteraciones:
    vertice = random.choice([v1, v2, v3])
    punto = (punto + vertice) / 2

    puntos_x.append(punto[0])
    puntos_y.append(punto[1])

    contador += 1
plt.figure(figsize=(6, 6))
plt.scatter(puntos_x, puntos_y, s=0.2, color="black")
plt.axis("off")
plt.title("Fractal de Sierpinski")
```
