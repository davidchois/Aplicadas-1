1. Encuentre las dimensiones del cilindro circular recto de volumen
máximo que puede inscribirse en un cono circular recto dado:

Solucion:

*Paso 1.* Definimos la función objetivo $Q$, que es el volumen del cilindro de radio $r$ y altura $h$:

$$
Q(r,h) \;=\;\pi\,r^2\,h.
$$



*Paso 2.* Usamos la geometría del cono de radio de base $R$ y altura $H$. Si apoyamos el cilindro sobre la base del cono, la sección transversal del cono a una altura $h$ sobre la base tiene radio

$$
r_{\text{cono}}(h)\;=\;R\Bigl(1 - \tfrac{h}{H}\Bigr).
$$

Para que el cilindro quepa justo dentro del cono, su radio debe coincidir con esa sección:

$$
r = R\Bigl(1 - \tfrac{h}{H}\Bigr).
$$

Sustituyendo en $Q$, queda como función de la sola variable $h$:

$$
Q(h)
= \pi \bigl[R(1 - \tfrac{h}{H})\bigr]^2\,h
= \pi R^2\,h\bigl(1 - \tfrac{h}{H}\bigr)^2.
$$


*Paso 3.* Buscamos los puntos críticos de $Q(h)$ en el intervalo $[0,H]$.

1. *Fronteras:*

   * $h=0$  $\Rightarrow Q(0)=0$.
   * $h=H$  $\Rightarrow Q(H)=0$.

2. *Puntos estacionarios:*

   $$
   Q(h)=\pi R^2\,h\,(1 - \tfrac{h}{H})^2.
   $$

   Derivamos respecto a $h$:

   $$
   Q'(h)
   = \pi R^2\Bigl[(1 - \tfrac{h}{H})^2 + h\cdot 2(1 - \tfrac{h}{H})\bigl(-\tfrac1H\bigr)\Bigr].
   $$

   Factorizamos:

   $$
   Q'(h)
   = \pi R^2\,(1 - \tfrac{h}{H})\Bigl[(1 - \tfrac{h}{H}) - \tfrac{2h}{H}\Bigr]
   $$

   $$
   = (\pi R^2)(1 - \tfrac{h}{H})(1 - \tfrac{3h}{H}).
   $$

   Igualamos a cero:

   $$
   (\pi R^2)(1 - \tfrac{h}{H})(1 - \tfrac{3h}{H}) = 0
   \quad\Longrightarrow\quad
   h = H
   \quad\text{o}\quad
   h = \tfrac{H}{3}.
   $$

3. *Punto singular:* no hay problemas de dominio interno aparte de $h\in[0,H]$.



*Paso 4.* Comprobamos cuál de estos puntos da el máximo:

* En $h=0$ y $h=H$, $Q=0$.
* En $h = \tfrac{H}{3}$,

  $$
  Q\bigl(\tfrac{H}{3}\bigr)
  = \pi R^2\;\tfrac{H}{3}\;\Bigl(1 - \tfrac{1}{3}\Bigr)^2
  = \pi R^2\;\tfrac{H}{3}\;\Bigl(\tfrac{2}{3}\Bigr)^2
  = \tfrac{4}{27}\,\pi R^2 H >0.
  $$

Además, la segunda derivada en $h=\tfrac H3$ resulta negativa, confirmando máximo.



*Conclusión.* El cilindro de volumen máximo inscrito tiene

$$
h = \frac{H}{3}
$$

$$ r = R\Bigl(1 - \tfrac{1}{3}\Bigr) = \frac{2R}{3} $$

2. Suponga que un pez nada río arriba con velocidad relativa al agua
$v$ y que la corriente del río tiene velocidad $-v_c$ (el signo negativo indica que la velocidad de la corriente es en dirección opuesta a la del pez). La energía empleada en recorrer una distancia $d$ a contracorriente es directamente proporcional al tiempo requerido para recorrer la distancia $d$ y el cubo de la velocidad. ¿Qué velocidad $v$ minimiza la energía empleada en nadar esta distancia?

Solucion:

**Paso 1.** Definimos la función objetivo $Q$, que representa la energía empleada por el pez para nadar río arriba una distancia $d$.  
Según el enunciado, la energía es proporcional al producto del **tiempo requerido** y el **cubo de la velocidad relativa al agua**.

Sabemos que:

- La velocidad del pez respecto al agua es $v$.
- La velocidad de la corriente es $-v_c$, por lo tanto, la **velocidad neta del pez respecto a la tierra** es $v - v_c$.
- El tiempo en recorrer la distancia $d$ es $t = \frac{d}{v - v_c}$.
- La energía es proporcional a $t \cdot v^3$.

Entonces, la función objetivo es:

$$
Q(v) = k \cdot \frac{d}{v - v_c} \cdot v^3
$$

Donde $k$ es la constante de proporcionalidad positiva.

**Paso 2.** Simplificamos la función $Q(v)$ como función de una sola variable $v$:

$$
Q(v) = k d \cdot \frac{v^3}{v - v_c}
$$

Dominio: $v > v_c$ (para que el pez avance río arriba).

**Paso 3.** Buscamos los puntos críticos de $Q(v)$ derivando la función:

Sea:

$$
Q(v) = k d \cdot \frac{v^3}{v - v_c}
$$

Derivamos respecto a $v$ usando la regla del cociente:

$$
Q'(v) = k d \cdot \left[ \frac{(3v^2)(v - v_c) - v^3(1)}{(v - v_c)^2} \right]
$$

$$
= k d \cdot \frac{3v^2(v - v_c) - v^3}{(v - v_c)^2}
$$

Igualamos a cero:

$$
3v^2(v - v_c) - v^3 = 0
$$

$$
v^2 \left( 3(v - v_c) - v \right) = 0
$$

$$
v^2 (2v - 3v_c) = 0
$$

Soluciones:

- $v = 0$ (no válida porque $v < v_c$ y el pez no avanza).
- $v = \frac{3}{2}v_c$

**Paso 4.** Verificamos que $v = \frac{3}{2}v_c$ da un mínimo.

Como el dominio es $v > v_c$ y la energía tiende a infinito cuando $v \to v_c^+$ o $v \to \infty$, el valor $v = \frac{3}{2}v_c$ es el **único punto interior** donde $Q'(v) = 0$.

Además, evaluando la segunda derivada o comparando con valores cercanos, se verifica que se trata de un **mínimo local absoluto**.

$$
\textbf{Conclusión:} \quad \text{La velocidad que minimiza la energía es:} \quad \boxed{v = \frac{3}{2}v_c}
$$

3. Un pasillo de 6 pies de ancho da vuelta en ángulo recto. ¿Cuál es
la longitud de la varilla delgada más larga que puede transportarse alrededor de la esquina, suponiendo que la varilla no puede doblarse?

Solucion:


**Paso 1.** Definimos la función objetivo $Q$, que es la longitud total de la varilla rígida formada por dos segmentos rectos que tocan ambas paredes del pasillo, uno en cada tramo del ángulo recto.

Sea $\theta$ el ángulo que forma el segmento horizontal de la varilla con la pared.

$$
Q(\theta) = \frac{6}{\sin(\theta)} + \frac{6}{\cos(\theta)}
$$

**Paso 2.** La función $Q$ ya está expresada como función de una sola variable $\theta$, con dominio $0 < \theta < \frac{\pi}{2}$.  
Buscamos minimizar esta función, ya que la mínima suma de trayectorias implica la varilla más larga posible.

**Paso 3.** Buscamos los puntos críticos de $Q(\theta)$ en el intervalo $(0, \frac{\pi}{2})$.

Fronteras:

$$
\lim_{\theta \to 0^+} Q(\theta) = \infty, \quad \lim_{\theta \to \frac{\pi}{2}^-} Q(\theta) = \infty
$$

Puntos estacionarios:

$$
Q(\theta) = \frac{6}{\sin(\theta)} + \frac{6}{\cos(\theta)}
$$

Derivamos respecto a $\theta$:

$$
Q'(\theta) = -6 \cdot \frac{\cos(\theta)}{\sin^2(\theta)} + 6 \cdot \frac{\sin(\theta)}{\cos^2(\theta)}
$$

Igualamos a cero:

$$
\frac{\sin(\theta)}{\cos^2(\theta)} = \frac{\cos(\theta)}{\sin^2(\theta)} \Rightarrow \sin^3(\theta) = \cos^3(\theta) \Rightarrow \tan(\theta) = 1 \Rightarrow \theta = \frac{\pi}{4}
$$

Punto singular: No hay problemas de dominio dentro del intervalo $(0, \frac{\pi}{2})$.

**Paso 4.** Comprobamos cuál de estos puntos da el mínimo:

En los extremos $\theta \to 0$ y $\theta \to \frac{\pi}{2}$, $Q \to \infty$

En $\theta = \frac{\pi}{4}$:

$$
\sin\left( \frac{\pi}{4} \right) = \cos\left( \frac{\pi}{4} \right) = \frac{\sqrt{2}}{2}
$$

$$
Q\left( \frac{\pi}{4} \right) = \frac{6}{\frac{\sqrt{2}}{2}} + \frac{6}{\frac{\sqrt{2}}{2}} = \frac{12}{\frac{\sqrt{2}}{2}} = \frac{24}{\sqrt{2}} = 12\sqrt{2}
$$

Además, la segunda derivada en $\theta = \frac{\pi}{4}$ resulta positiva, confirmando mínimo.

$$
\textbf{Conclusión:} \quad \text{La varilla más larga que puede transportarse por la esquina de un pasillo de 6 pies de ancho es:} \quad \boxed{L = 12\sqrt{2} \approx 16.97 \text{ pies}}
$$

Un objeto que se arroja desde el borde de un acantilado de 100 pies, sigue la trayectoria dada por $
h(x) = -\frac{x^2}{100} + x + 100
$. Un observador se encuentra parado a 2 pies del fondo del acantilado. (a) Encuentre la posición del objeto cuando está más cerca del observador. (b) Encuentre la posición del objeto cuando está más lejos del observador.

Sea el objeto en $(x,h(x))$ con

$$
h(x) \;=\; -\frac{x^2}{100} \;+\; x \;+\;100,
$$

y el observador en el pie del acantilado, a una altura de 2 ft, es decir en $(0,2)$. La distancia euclídea entre ambos es

$$
D(x) \;=\;\sqrt{\bigl(x-0\bigr)^2 \;+\;\bigl(h(x)-2\bigr)^2},
$$

y para simplificar el análisis usamos

$$
Q(x) \;=\; D^2(x)
\;=\;x^2 \;+\;\bigl(h(x)-2\bigr)^2.
$$

El dominio físico va de $x=0$ (lanzamiento) hasta el punto donde $h(x)=0$, i.e.\\

$$
-\tfrac{x^2}{100}+x+100=0
\;\Longrightarrow\;
x = 50\pm 50\sqrt5,
$$

y tomamos la solución positiva
$\;x_{\max} = 50+50\sqrt5\approx161.803$.

\

**Puntos críticos**

* **Fronteras:**

  $$
    Q(0) = 0^2 + (100-2)^2 = 98^2 = 9604,
    \qquad
    Q(50+50\sqrt5) = (50+50\sqrt5)^2 + (0-2)^2 \approx 161.803^2 + 4.
  $$
* **Estacionarios:**

  $$
    Q(x)=x^2 + \Bigl(-\tfrac{x^2}{100}+x+98\Bigr)^2,
  $$

  $$
    Q'(x)
    =2x \;+\;2\bigl(h(x)-2\bigr)\bigl(h'(x)\bigr),
    \quad
    h'(x)=-\tfrac{2x}{100}+1.
  $$

  Al resolver $Q'(x)=0$ en $(0,161.803)$ se obtiene **ninguna raíz real** en el intervalo, de modo que no hay puntos estacionarios interiores.
* **Singularidades:**
  $Q$ es continua y derivable en $[0,161.803]$, salvo en los extremos (donde se evalúa por frontera).

\

**Comparación de valores**

$$
Q(0)=9604,
\qquad
Q(161.803)\approx161.803^2+4\approx26188,
$$

y no hay ningún otro candidato.

\

**Conclusión**

* El **mínimo** de la distancia (el objeto más cerca del observador) se da al **momento del lanzamiento**, en

  $$
    x=0,\quad h(0)=100.
  $$
* El **máximo** de la distancia (el objeto más lejos) se da al **impactar el suelo**, en

  $$
    x=50 + 50\sqrt5\approx161.803,\quad h=0.
  $$

\

**Respuesta final:**

$$
\text{(a) La posición más cercana es }(x,h)=(0,100)
$$

$$
\text{(b) la más lejana es }(x,h)=\bigl(50+50\sqrt5,\;0\bigr).
$$

## Análisis de la función $f(x) = \frac{x}{x+1}$ ##

### Dominio: ###
El denominador se anula cuando $x + 1 = 0 \Rightarrow x = -1$.  
Por lo tanto, el dominio es $\mathbb{R} \setminus \{-1\}$.

### Rango: ###
Igualamos $y = \frac{x}{x+1}$ y despejamos:

$$
y(x + 1) = x \Rightarrow yx + y = x \Rightarrow y = x(1 - y) \Rightarrow x = \frac{y}{1 - y}
$$

La expresión no está definida si $1 - y = 0 \Rightarrow y = 1$.  
Entonces, el rango es $\mathbb{R} \setminus \{1\}$.

### Intersecciones: ###
- Con el eje $x$: $f(x) = 0 \Rightarrow x = 0 \Rightarrow (0, 0)$  
- Con el eje $y$: $f(0) = 0 \Rightarrow (0, 0)$

### Simetría: ###
Se evalúa $f(-x) = \frac{-x}{1 - x}$.  
Esto no es igual a $f(x)$ ni a $-f(x)$, por lo tanto, la función no es par ni impar.

### Asíntotas: ###

#### Asíntota vertical: ####
Existe una asíntota vertical en $x = -1$, donde el denominador se anula.

#### Asíntota horizontal : ####
- El numerador tiene grado 1 y coeficiente principal 1.  
- El denominador tiene grado 1 y coeficiente principal 1.  

Como los grados son iguales, la asíntota horizontal se calcula como:

$$
y = \frac{1}{1} = 1
$$

Por lo tanto, hay una asíntota horizontal en $y = 1$.

### Primera derivada: ###
Aplicamos la regla del cociente:

$$
f'(x) = \frac{(x + 1)(1) - x(1)}{(x + 1)^2} = \frac{1}{(x + 1)^2}
$$

Como $f'(x) > 0$ para todo $x \neq -1$, la función es estrictamente creciente en todo su dominio.

### Segunda derivada: ###
Derivamos nuevamente:

$$
f''(x) = \frac{d}{dx} \left( \frac{1}{(x + 1)^2} \right) = \frac{-2}{(x + 1)^3}
$$

- Si $x < -1$, entonces $f''(x) > 0$ ⇒ cóncava hacia arriba  
- Si $x > -1$, entonces $f''(x) < 0$ ⇒ cóncava hacia abajo  
- En $x = -1$, $f''(x)$ no está definida ⇒ no hay punto de inflexión

## Resumen general: ##
- Dominio: $\mathbb{R} - \{-1\}$
- Rango: $\mathbb{R} - \{1\}$
- Intersección con los ejes: $(0, 0)$
- Asíntota vertical: $x = -1$
- Asíntota horizontal: $y = 1$
- La función es estrictamente creciente en todo su dominio
- Concavidad:
- Cóncava hacia arriba en $(-\infty, -1)$
- Cóncava hacia abajo en $(-1, \infty)$
- No tiene puntos críticos
- No tiene puntos de inflexión
- No es par ni impar

## Análisis de la función $g(x) = \frac{|x| + x}{2}(3x + 2)$ ##

### Dominio: ###
La función está compuesta por valor absoluto y operaciones algebraicas elementales.  
Está definida para todo número real.  
Resultado: $\mathbb{R}$

### Simetría: ###
Evaluamos:

$$
g(-x) = \frac{|-x| + (-x)}{2}(3(-x) + 2) = \left( \frac{x - x}{2} \right)(-3x + 2) = 0
$$

$$
g(x) = x(3x + 2) \quad \text{para } x \geq 0
$$

Como $g(-x) \neq g(x)$ y $g(-x) \neq -g(x)$,  
Resultado: no es par ni impar

### Rango: ###
Por tramos:

- Si $x < 0$:
  $$
  g(x) = 0
  $$

- Si $x \geq 0$:
  $$
  g(x) = x(3x + 2) = 3x^2 + 2x
  $$

Buscamos el mínimo de $g(x)$ en $x \geq 0$:
- $g(0) = 0$
- Como $g(x)$ es un polinomio cuadrático con coeficiente positivo, crece hacia $\infty$

Resultado: $[0, \infty)$

### Asíntotas horizontales: ###

#### Método por grados: ####
- En $x < 0$: $g(x) = 0$ → se aplana sobre $y = 0$  
- En $x \geq 0$: $g(x) = 3x^2 + 2x$ (grado 2) → tiende a infinito

Resultado: asíntota horizontal en $y = 0$ cuando $x \to -\infty$

### Derivada primera por tramos: ###

- Si $x < 0$:
  $$
  g(x) = 0 \Rightarrow g'(x) = 0
  $$

- Si $x \geq 0$:
  $$
  g(x) = x(3x + 2)
  $$

Aplicamos derivada del producto:

$$
g'(x) = \frac{d}{dx}(x) \cdot (3x + 2) + x \cdot \frac{d}{dx}(3x + 2) = 1 \cdot (3x + 2) + x \cdot 3 = 3x + 2 + 3x = 6x + 2
$$

Resultado:

$$
g'(x) =
\begin{cases}
0 & \text{si } x < 0 \\
6x + 2 & \text{si } x \geq 0
\end{cases}
$$

### Puntos críticos: ###

Resolvemos:

$$
6x + 2 = 0 \Rightarrow x = -\frac{1}{3}
$$

### Segunda derivada por tramos: ###

- Si $x < 0$:
  $$
  g''(x) = 0
  $$

- Si $x \geq 0$:
  $$
  g''(x) = \frac{d}{dx}(6x + 2) = 6
  $$

Resultado:

$$
g''(x) =
\begin{cases}
0 & \text{si } x < 0 \\
6 & \text{si } x \geq 0
\end{cases}
$$

### Verificación de puntos críticos: ###

$$
g''(-\tfrac{1}{3}) = 6 > 0 \Rightarrow \text{mínimo local}
$$

### Puntos de inflexión: ###

Buscamos:

$$
g''(x) = 0 \Rightarrow \text{no hay solución} \Rightarrow \text{no hay punto de inflexión}
$$



## Resumen general: ##
- Dominio: $\mathbb{R}$
- Rango: $[0, \infty)$
- Intersección: $(0, 0)$
- Asíntota horizontal: $y = 0$ para $x \to -\infty$
- Crecimiento:
- Constante en $x < 0$
- Creciente en $x > 0$
- Punto crítico: $x = -\frac{1}{3}$, mínimo local
- Concavidad:
- Sin curvatura en $x < 0$ (función constante)
- Cóncava hacia arriba en $x > 0$
- Puntos de inflexión: ninguno
- Simetría: no es par ni impar

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función y su derivada para el Ejercicio I.1

def f1(x):
    return x**5 - 6*x**3 + 2*x - 2

def f1_prime(x):
    return 5*x**4 - 18*x**2 + 2

# Generar valores y graficar
x_vals_f1 = np.linspace(-3, 3, 400)
y_vals_f1 = f1(x_vals_f1)

plt.figure(figsize=(10, 6))
plt.plot(x_vals_f1, y_vals_f1, label='$f(x) = x^5 - 6x^3 + 2x - 2$')

plt.axhline(0, color='black', lw=0.5)
plt.title('Gráfica de $f(x) = x^5 - 6x^3 + 2x - 2$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
def newton_raphson(f, fprime, x0, tol, maxiter):
    x = x0
    for i in range(maxiter):
        fx = f(x)
        if abs(fx) < tol:
            return x

        fp = fprime(x)
        if fp == 0:
            print('Error la derivada es cero')
            break
        x = x - fx / fp
    return x

# Ejemplo de uso:
f = lambda x: (x**5) - (6*x**3) + 2*x - 2
g = lambda x: 5*x**4 - 18*x**2 + 2

x0 = -2
tol = 1e-5
maxiter = 10000
root = newton_raphson(f, g, x0, tol, maxiter)
print(f"La raíz de la función es: {root}")

suposiciones_iniciales = [-2.4, -0.7, 2.5]
roots_f1 = []
tolerance_f1 = 0.0001
max_iterations_f1 = 10000

print("Resolviendo para f(x) = x^5 - 6x^3 + 2x - 2 = 0:")
for guess in suposiciones_iniciales:
    print(f"Intento con valor inicial: x0 = {guess}")
    root = newton_raphson(f1, f1_prime, guess, tol=tolerance_f1, maxiter=max_iterations_f1)

    val_f_at_root = f1(root) # Siempre se calcula f(root) para verificar
    if abs(val_f_at_root) < tolerance_f1:
        if not any(abs(r_val - root) < 1e-5 for r_val in roots_f1): # Evitar duplicados
            roots_f1.append(root)
        print(f"Convergencia exitosa. Raíz encontrada: {root:.7f}, f({root:.7f}) = {val_f_at_root:.7f}\n")
    else:
        print(f"Proceso terminó. Valor devuelto: {root:.7f}, f({root:.7f}) = {val_f_at_root:.7f}. No cumple |f(x)| < {tolerance_f1}.\n")

print("Las raíces aproximadas para x^5 - 6x^3 + 2x - 2 = 0 (que cumplen |f(x)| < 0.0001) son:")
roots_f1.sort()
for r_idx, r_val in enumerate(roots_f1):
    print(f"x_{r_idx+1} approx {r_val:.7f}")
```
